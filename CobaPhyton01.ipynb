```python
#import Libary Matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Import Libary Numpy
import numpy as np

#Menghitung jumlah data 100 yang berjarak dari 0 sampai 2n
x = np.linspace(0,2*np.pi, 100)
#Kemudian menghitung cos si x
cos_x = np.cos(x)
```

```python
#membuat line plot
flg, ax = plt.subplots()
_ = ax.plot(x, cos_x)
```

```python
#Membuat sumbu x dan y memiliki rasio yang sama
fig, ax = plt.subplots()
_ = ax.plot(x, cos_x)
_ = ax.set_aspect('equal')
```

```python
fig, ax = plt.subplots()
_ = ax.plot(x, cos_x, markersize=30, linestyle='-.',
            color = 'yellow', label='cos')
_ = ax.set_aspect('equal')
_ = ax.legend()
```
